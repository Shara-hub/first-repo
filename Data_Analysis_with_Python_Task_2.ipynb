```python
import pandas as pd

# Assuming 'data2' is your DataFrame
# Select only numerical columns for quantile calculation
numerical_data2 = data2.select_dtypes(include=['number'])

# Calculate quantiles only on numerical columns
Q1 = numerical_data2.quantile(0.25)
Q3 = numerical_data2.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data2=numerical_data2[~((numerical_data2<(Q1-1.5*IQR))|(numerical_data2>(Q3+1.5*IQR))).any(axis=1)]
data2
```
